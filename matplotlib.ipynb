```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [33,76,54,23,62,75,45,85,78]    
```

```python
plt.plot(x,y,color ='green') # colour = used to change the color
```

```python
 plt.plot(x,y,color ='green', linewidth = 5)  # line =width is used to increasde the width of line 
```

```python
 plt.plot(x,y,color ='green', linewidth = 5, linestyle = "dotted") #linestyle is used to change the style of line 
```

```python
plt.xlabel ('day')
plt.ylabel('wind speed')
plt.title('Weather')
plt.plot(x,y,color ='green', linewidth = 5, linestyle = "dotted")
```
