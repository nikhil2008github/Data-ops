```python
days = [1,2,3,4,5,6,7]
max_temp = [34,56,53,45,55,44,48]
min_temp =[22,34,25,38,26,32,31]
avg_temp = [30,37,45,36,50,40,36]
```

#**BAR CHART**

```python
company = ['google','Amazon','msft','fb']
revenue= [90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
#plt.xticks (ypos,revenue) is used to label of particular bar by numpy 
plt.xticks([0,1,2,3],['google','Amazon','msft','fb'],rotation=0)
plt.bar(xpos,revenue)
```

```python
#plt.xticks (ypos,revenue) is used to label of particular bar by numpy 
# plt.xlabel()
plt.ylabel('Revenue(blln)')
plt.title('US Tech stocks')
plt.xticks([0,1,2,3],['google','Amazon','msft','fb'],rotation=0)
plt.bar(xpos,revenue)
```

```python
#plt.xticks (ypos,revenue) is used to label of particular bar by numpy 
# plt.xlabel()
plt.ylabel('Revenue(blln)')
plt.title('US Tech stocks')
plt.xticks([0,1,2,3],['google','Amazon','msft','fb'],rotation=0)
plt.bar(xpos- 0.2,revenue,width =0.4, label= 'Revenue')
plt.bar(xpos + 0.2,profit,width = 0.4,label='profit')
plt.legend()
```

```python
#plt.xticks (ypos,revenue) is used to label of particular bar by numpy 
# plt.xlabel()
plt.ylabel('Revenue(blln)')
plt.title('US Tech stocks')
plt.yticks([0,1,2,3],['google','Amazon','msft','fb'],rotation=0)
plt.barh(xpos- 0.2,revenue, label= 'Revenue')
plt.barh(xpos + 0.2,profit,label='profit')
plt.legend()
```
