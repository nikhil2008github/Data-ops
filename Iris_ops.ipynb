```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'/content/Iris.csv')
df
```

```python
count_setosa = 0
for i in df["Species"]:
  if i == "Iris-setosa":

      count_setosa = count_setosa + 1

print(count_setosa)
```

```python
count_vir = 0
for i in df["Species"]:
  if i == "Iris-virginica":

      count_vir = count_vir + 1

print(count_vir)
```

```python
df["Species"].unique()
```

```python
count_ver = 0
for i in df["Species"]:
  if i == "Iris-versicolor":

      count_ver = count_ver + 1

print(count_ver)
```

```python
import matplotlib.pyplot as plt
```

```python
data = [count_setosa, count_vir, count_ver]
labels = ["setosa","virginica","versicolor"]


plt.pie(x = data, labels = labels)
plt.show()
```
